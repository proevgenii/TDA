```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.tri as tri
from sklearn.neighbors import KDTree
%matplotlib inline 
```

### Функция генерации точек на диске по равномерному распределению

```python
def sample(r_outer, r_inner):
    rad = math.sqrt(random.uniform(r_inner**2, r_outer**2))
    angle = random.uniform(-math.pi, math.pi)
    x, y = rad*math.cos(angle),rad*math.sin(angle)
    return x, y
```

```python
x, y = [], []
for i in range(1000):
  xs, ys = random.uniform(-10,10), random.uniform(-10,10)
  x.append(xs)
  y.append(ys)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x, y)
None
```

### Отрисовка графиков 

```python
fig, ax = plt.subplots(3,4, figsize=(32,24))

for k in range(3):
  index = []
  samples = [sample(10, max(2, k*3)) for _ in range(100)]
  xs, ys = zip(*samples)
  X = np.array(samples)
  tree = KDTree(X, leaf_size=3, metric='euclidean', )
  for knn in (3, 5, 10 , 40):
    dist, ind = tree.query(X, knn) 
    ind = [[ind[0], ind[-2], ind[-1]] for ind in ind ]
    index.append(ind)

  for i in range(4):
    ax[k][i].add_patch(plt.Circle((0, 0), 10, color='b', fill=False))
    ax[k][i].add_patch(plt.Circle((0, 0), max(1, k*3), color='b', fill=False))
    ax[k][i].scatter(xs, ys, s=500, marker='.')
    ax[k][i].axis("equal")
    triang = tri.Triangulation(xs, ys, triangles=index[i])
    ax[k][i].triplot(triang, "g-", alpha=.4)
  
  
plt.show()
```

### Функция генерации точек внутри лемнискаты по равномерному распределению

```python
def sample_lem(alpha, alpha1):
  xs =[]
  ys =[]
  alpha1 =alpha1
  alpha = alpha
  while len(xs) < 100:
    for t in np.linspace(0, 2*np.pi, num=100):
      x = alpha * np.sqrt(2) * np.cos(t) / (np.sin(t)**2 + 1)
      y = alpha * np.sqrt(2) * np.cos(t) * np.sin(t) / (np.sin(t)**2 + 1)
      
      x1 = alpha1 * np.sqrt(2) * np.cos(t) / (np.sin(t)**2 + 1)
      y1 = alpha1 * np.sqrt(2) * np.cos(t) * np.sin(t) / (np.sin(t)**2 + 1)

      x_g, y_g, = random.uniform(-10, 10), random.uniform(-3, 3)

      if ((x1<x_g<x and y>y_g>y1)
       or (x1<x_g<x and y<y_g<y1)
        or (x1>x_g>x and y>y_g>y1)
         or (x1>x_g>x and y<y_g<y1)):
        xs.append(x_g)
        ys.append(y_g)
  return xs, ys
```

```python
fig, ax = plt.subplots(3,4, figsize=(32,18))

for k in range(3):
  for i in range(4):

    #Внешняя 8-ка
    alpha = 7.1
    t = np.linspace(0, 2*np.pi, num=100)
    x = alpha * np.sqrt(2) * np.cos(t) / (np.sin(t)**2 + 1)
    y = alpha * np.sqrt(2) * np.cos(t) * np.sin(t) / (np.sin(t)**2 + 1)
    ax[k][i].plot(x, y)

    # Внутренняя 8-ка
    alpha1 = max(1, k*3)
    x1 = alpha1 * np.sqrt(2) * np.cos(t) / (np.sin(t)**2 + 1)
    y1 = alpha1 * np.sqrt(2) * np.cos(t) * np.sin(t) / (np.sin(t)**2 + 1)
    ax[k][i].plot(x1, y1)

    # Точки
    xs, ys = sample_lem(alpha=alpha, alpha1=alpha1)
    ax[k][i].scatter(xs, ys, s=100, marker='.')

    # Триангуляция
    X = np.stack((xs, ys), axis=1)
    tree = KDTree(X, leaf_size=3, metric='euclidean', )
    for knn in (3, 5, 10 , 40):
      dist, ind = tree.query(X, knn) 
      ind = [[ind[0], ind[-2], ind[-1]] for ind in ind ]
      index.append(ind)

    ax[k][i].axis("equal")
    triang = tri.Triangulation(xs, ys, triangles=index[i])
    ax[k][i].triplot(triang, "g-", alpha=.4)

plt.show()
```

```python

import kmapper as km

# Some sample data
from sklearn import datasets
data, labels = datasets.make_circles(n_samples=5000, noise=0.03, factor=0.3)

# Initialize
mapper = km.KeplerMapper(verbose=1)

# Fit to and transform the data
projected_data = mapper.fit_transform(data, projection=[0,1]) # X-Y axis

# Create dictionary called 'graph' with nodes, edges and meta-information
graph = mapper.map(projected_data, data, cover=km.Cover(n_cubes=10))

# Visualize it
mapper.visualize(graph, path_html="make_circles_keplermapper_output.html",
                 title="make_circles(n_samples=5000, noise=0.03, factor=0.3)")
```
